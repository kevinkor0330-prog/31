```python
import glob
import pandas as pd

files = glob.glob(r"C:\Users\user\Desktop\python_lab\psql_insert\홈택스_거래처목록\*.xls")

dfs = [pd.read_excel(f, engine="xlrd", header=3) for f in files]
df = pd.concat(dfs, ignore_index=True).reset_index(drop = True)

df["순번"] = df["순번"].astype(str).str.replace(",", "").astype(float).astype("Int64")
df = df.sort_values("순번")

df = df[["순번", "거래처등록번호", "거래처상호", "사업자주소","업태", "종목", "이메일주소","구분","등록일자"]]

df_unique_biz_code = df.drop_duplicates(subset = "거래처등록번호")

df_unique_biz_code['거래처등록번호'] = df_unique_biz_code['거래처등록번호'].apply(
    lambda x: x[:7] + '*******' if len(str(x)) == 14 else x
)

df_unique_biz_code.to_excel("홈택스_거래처.xlsx")
```
