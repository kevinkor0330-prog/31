```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
import glob
 
 # 월별, 거래처
# --- Option 1: Using glob to auto-detect all matching .xls files ---
all_files = glob.glob('*.xls')  # Finds all matching files automatically
 
 
all_data = []
 
for file in all_files:
    df = pd.read_excel(file, sheet_name='출하자료')
    df = df[df['일자'].notna()]
    
    column_i_need = df[['일자', '거래처', '납품처', '출하중량(Kg)', '매출액', '제품군']]
    column_i_need = column_i_need.copy()  # Avoid SettingWithCopyWarning
    column_i_need['일자'] = column_i_need['일자'].dt.strftime('%Y-%m')
    
    grouped = column_i_need.groupby(['일자', '거래처'], as_index=False).agg({
        '납품처': 'unique',
        '출하중량(Kg)': 'sum',
        '매출액': 'sum',
        '제품군': 'first'
    })
    

    all_data.append(grouped)
 
final_df = pd.concat(all_data, ignore_index=True)
final_df
```

```python
sorted = final_df.sort_values('매출액', ascending= False)
sorted
```

```python
 # 그냥 거래처 그룹바이
# --- Option 1: Using glob to auto-detect all matching .xls files ---
all_files = glob.glob('*.xls')  # Finds all matching files automatically
 
 
all_data = []
 
for file in all_files:
    df = pd.read_excel(file, sheet_name='출하자료')
    df = df[df['일자'].notna()]
    
    column_i_need = df[['일자', '거래처', '납품처', '출하중량(Kg)', '매출액', '제품군']]
    column_i_need = column_i_need.copy()  # Avoid SettingWithCopyWarning
    column_i_need['일자'] = column_i_need['일자'].dt.strftime('%Y-%m')


    all_data.append(column_i_need)
 
original_concat = pd.concat(all_data, ignore_index=True)
original_concat
```

```python
grouped = original_concat.groupby(['거래처'], as_index=False).agg({
    '납품처': lambda x: ', '.join(filter(None, x.unique())),
    '출하중량(Kg)': 'sum',
    '매출액': 'sum',
    '제품군': 'first'
})
```

```python
grouped.sort_values('매출액', ascending= False).to_excel("2025~2026년 매출통합자료.xlsx")
```

```python
grouped['납품처'] = grouped['납품처'].astype(str)
grouped.to_excel("2025~2026년 매출통합자료.xlsx")
```

```python
# 1. Setup
plt.rc('font', family='Malgun Gothic') 
final_df['Label'] = final_df['일자'].astype(str) + " | " + final_df['거래처']
df_plot = final_df.sort_values('매출액', ascending=False)

# 2. Dynamically calculate height
# We want about 0.3 inches per row. If you have 100 rows, height = 30 inches.
dynamic_height = len(df_plot) * 0.3 

# 3. Plot
plt.figure(figsize=(12, dynamic_height)) # Width is 12, Height scales with data
sns.barplot(data=df_plot, x='매출액', y='Label', palette='magma')

plt.title('All Sales Events (Date + Partner)', fontsize=16)
plt.xlabel('Sales Amount (매출액)')
plt.ylabel('Date | Partner')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# 4. Save with High Quality
# bbox_inches='tight' ensures labels aren't cut off at the edges
plt.savefig('all_sales_report.jpg', format='jpg', dpi=300, bbox_inches='tight')

print(f"Saved {len(df_plot)} rows to all_sales_report.jpg")
plt.show()
```
