```python
import numpy as np
```

```python
import pandas as pd

jaehak_df = pd.read_excel("단가인상_현황_취합_og_최재학.xlsx", header =5).dropna(subset= '거채처')
jaehak_df = jaehak_df.drop(columns = 'Unnamed: 0')
jaehak_df = jaehak_df[['거채처', '담당자', '1~2월 매출', '1~2월 매출중량\n(TON)','요청\n\/g','예상\n\/g', '합의\n\/g', '적용\n시점', '진행상황']]
jaehak_df = jaehak_df.rename(columns = {'거채처':'거래처'})
jaehak_df
```

```python
import msoffcrypto
import io

file_path = '00_영업실적_2024년~.xlsx'
password = '3101'

# 1. 암호화된 파일 열기
temp_file = io.BytesIO()
with open(file_path, 'rb') as f:
    office_file = msoffcrypto.OfficeFile(f)
    # 비밀번호 입력하여 해독
    office_file.load_key(password=password)
    office_file.decrypt(temp_file)
temp_file.seek(0)
마감자료 = pd.read_excel(temp_file, engine='openpyxl')
마감자료 = 마감자료.dropna(subset= '년도')
마감자료 = 마감자료[['년도', '월', '대분류','합계(운임비/금형비+공급가액)', '중량(Ton)']]
filtered_data = 마감자료[(마감자료['년도'] == 26) & (마감자료['월'] == 3)]
filtered_data = filtered_data.sort_values('중량(Ton)', ascending= False).reset_index(drop = True)
filtered_data
```

```python
#코웨이 가화텍 필터

filtered_data['대분류'] = filtered_data['대분류'].str.replace('^코웨이.*', '코웨이', regex=True)
filtered_data['대분류'] = filtered_data['대분류'].str.replace('^가화텍.*', '가화텍', regex=True)
result = filtered_data.groupby('대분류', as_index=False).sum(numeric_only=True)
result = result.sort_values("중량(Ton)", ascending= False).reset_index(drop= True).drop(columns = ['년도', '월']).rename(columns ={'대분류': '거래처'})
result
```

```python
#현재 단가인상자료와 merge
merged = pd.merge(jaehak_df, result, on = '거래처', how = 'outer', indicator= True).fillna(0)   
merged['1~3월매출'] = merged['1~2월 매출'] + merged['합계(운임비/금형비+공급가액)']
merged['1~3월매출중량(TON)'] = merged['1~2월 매출중량\n(TON)'] +  merged['중량(Ton)']
merged
```

```python
merged = merged.sort_values("1~3월매출중량(TON)", ascending= False).reset_index(drop= True)
```

```python
merged_final = merged[['거래처', '1~3월매출', '1~3월매출중량(TON)', '요청\n\/g', '예상\n\/g', '합의\n\/g', '적용\n시점', '진행상황','_merge']]
```

```python
# 적용시점 데이트 바꾸는중..
merged_final['적용\n시점'] = merged_final['적용\n시점'].replace(0,'null')
# 1. Convert to datetime (if it isn't already)
merged_final['적용\n시점'] = pd.to_datetime(merged_final['적용\n시점'], errors='coerce')

# 2. Format to m/d. 
# Note: Use %m/%d for "04/15" or %#m/%#d for "4/15" on Windows.
merged_final['적용\n시점'] = merged_final['적용\n시점'].dt.strftime('%m/%d').fillna('')
```

```python
merged_final['진행상황'] =merged_final['진행상황'].replace(0,'')
```

```python
merged_final['합의\n\/g'] = merged_final['합의\n\/g'].replace(0,'')
```
